### Membersihkan dataset orders

Dataset orders merupakan datset yang paling penting, karena didalamnya berisi data penting tentang kegiatam e-commerce. Pada analisis data kali ini, data yang akan digunakan hanya data orders yang sudah dikirimkan

```python
# Melihat jumlah order berdasar orer_status
print(orders["order_status"].value_counts())
```

```python
# Menghitung jumlah data jika data yang diambil hanya data  delivered saja
delivered_percentage = (orders["order_status"].value_counts()["delivered"] / len(orders)) * 100
print("Jika hanya data delivered yang diambil, data yang diperoleh sebanyak " + str(round(delivered_percentage, 2)) + "% of data")
```

Data yang diperoleh sebanyak 97.02% yang merupakan jumlah yang masih besar. Sehingga data orders yang akan digunakan untuk analisis lebih lanjut adalah data orders "delivered". Data lain akan di drop

```python
# Menghapus data yang belum atau yang tidak terkirim
delivered_orders = orders.drop(orders[orders['order_status'] != 'delivered'].index)
orders = delivered_orders
orders.head()
```

Memeriksa kembai missing value dataset orders

```python
orders.isna().sum()
```

Dapat disimpulkan bahwa kolom order_approved_at dan order_delivered_carrier_date tidak terlalu berpengaruh pada pertanyaan yang sudah diajukan. Akan tetapi terdapat 8 data yang null pada kolom order_delivered_customer_date yang berisi data tanggal dimana order diterima oleh customer. Data tersebut akan di drop

```python
# Drop data null pada order_approved_at, order_delivered_carrier_date dan order_delivered_customer_date
orders = orders.dropna(subset=["order_approved_at", "order_delivered_carrier_date", "order_delivered_customer_date"])
orders.isna().sum()
```

Menambahkan satu kolom untuk melihat order tepat waktu atau tidak. Tapi sebelum itu, harus mengubah data type yang masih berupa object



```python
datetime_columns = ["order_purchase_timestamp", "order_approved_at", "order_delivered_carrier_date", "order_delivered_customer_date", "order_estimated_delivery_date"]
for column in datetime_columns:
  orders[column] = pd.to_datetime(orders[column])
orders.info()
```

```python
orders["delivered_on_time"] = np.where(orders["order_delivered_customer_date"] <= orders["order_estimated_delivery_date"], "On Time", "Late")
```

```python
orders.sample(5)
```

## Visualization & Explanatory Analysis

### Pertanyaan 1: Bagaimana grafik penjualan di setiap bulanya?

Perlu membuat DataFrame baru yang menampung informasi jumlah order dan total revene per bulan untuk menjawab pertanyaan ini. Sehingga frekuensi data yang awalnya harian harus diubah menjadi bulanan. Kemudian informasi tentang performa penjualan dan revenue akan divisualisasikan menggunakan line chart

```python
monthly_orders = all_data.resample(rule='ME', on='order_approved_at').agg({
    "order_id": "nunique",
    "payment_value": "sum"
})
monthly_orders.index = monthly_orders.index.strftime('%B')
monthly_orders = monthly_orders.reset_index()
monthly_orders.rename(columns={
    "order_approved_at": "order_date",
    "order_id": "order_count",
    "payment_value": "revenue"
}, inplace=True)
monthly_orders
```

```python
# Menghapus data duplikasi per bulan dan hanya mengambil data terakhir
monthly_orders=monthly_orders.sort_values("order_count").drop_duplicates("order_date", keep='last')
```

```python
monthly_orders.sort_values(by="order_count", ascending=False)
```

```python
months_order = ["January", "February", "March", "April", "May", "June", "July", "August", "September", "October", "November", "December"]
monthly_orders["order_date"] = pd.Categorical(monthly_orders["order_date"], categories=months_order, ordered=True)
monthly_orders = monthly_orders.sort_values("order_date")

plt.figure(figsize=(15,7))
plt.plot(monthly_orders["order_date"], monthly_orders["order_count"], marker="o", linewidth=4)
plt.title("Jumlah Pesanan per Bulan (2018)", loc="center", fontsize=14)
plt.xlabel("Month", fontsize=12)
plt.ylabel("Jumlah Pesanan", fontsize=12)
plt.xticks(rotation=45, fontsize=10)
plt.yticks(fontsize=10)
plt.show()
```

Terdapat penurunan signifikan pada bulan Agustus dan terjadi kenaikan Oktober

### Pertanyaan 2: Diantara produk yang ada, yang manakah yang paling tertinggi dan yang paling rendah terjual?

Membuat DataFrame baru yang berisikan data produk dan kuantitas. Lalu informasi jumlah produk terjual akan divisualisasikan dengan bar chart

```python
sum_order_items = all_data.groupby("product_category_name_english")["product_id"].count().reset_index()
sum_order_items = sum_order_items.rename(columns={"product_id": "products"})
sum_order_items = sum_order_items.sort_values(by="products", ascending=False)
sum_order_items
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(24, 6))

colors = ["#068DA9", "#D3D3D3", "#D3D3D3", "#D3D3D3", "#D3D3D3"]

sns.barplot(x="products", y="product_category_name_english", data=sum_order_items.head(5), palette=colors, ax=ax[0])
ax[0].set_ylabel(None)
ax[0].set_xlabel(None)
ax[0].set_title("Produk paling banyak terjual", loc="center", fontsize=18)
ax[0].tick_params(axis ='y', labelsize=15)

sns.barplot(x="products", y="product_category_name_english", data=sum_order_items.sort_values(by="products", ascending=True).head(5), palette=colors, ax=ax[1])
ax[1].set_ylabel(None)
ax[1].set_xlabel(None)
ax[1].invert_xaxis()
ax[1].yaxis.set_label_position("right")
ax[1].yaxis.tick_right()
ax[1].set_title("Produk paling sedikit terjual", loc="center", fontsize=18)
ax[1].tick_params(axis='y', labelsize=15)

plt.suptitle("Produk paling banyak dan paling sedikit terjual", fontsize=20)
plt.show()
```

Dari bar chart diatas terlihat bahwa produk yang paling banyak terjual adalah bed_bath_table dan produk paling sedikit terjual adalah security_and_services

### Pertanyaan 3: Bagaimana jumlah pelanggan berdasar kota dan state?

- Demografi pelanggan berdasar kota

```python
bycity = all_data['customer_city'].value_counts().head(10)

plt.figure(figsize=(12, 6))

most_common_city = bycity.idxmax()

bycity = bycity.sort_values(ascending=False)

sns.barplot(x=bycity.index,
            y=bycity.values,
            palette=["#068DA9" if city == most_common_city else "#D3D3D3" for city in bycity.index]
            )

plt.title("Jumlah Pelanggan Berdasarkan Kota", fontsize=15)
plt.xlabel("Kota")
plt.ylabel("Jumlah Pelanggan")
plt.xticks(rotation=45, fontsize=10)
plt.show()
```

- Demografi berdasar state

```python
bystate = all_data['customer_state'].value_counts().head(10)

plt.figure(figsize=(12, 6))

most_common_city = bystate.idxmax()

bystate = bystate.sort_values(ascending=False)

sns.barplot(x=bystate.index,
            y=bystate.values,
            palette=["#068DA9" if city == most_common_city else "#D3D3D3" for city in bystate.index]
            )

plt.title("Jumlah Pelanggan Berdasarkan State", fontsize=15)
plt.xlabel("State")
plt.ylabel("Jumlah Pelanggan")
plt.xticks(rotation=45, fontsize=10)
plt.show()
```
